```python
#Loading the data

import os
import pandas as pd

df= pd.read_csv('audible_uncleaned.csv')
print(df.head)
```

```python
# getting the columns of the dataset
columns = list(df.columns)
columns
```

```python
# examining missing values
print("Missing values distribution: ")
print(df.isnull().mean())
print("")
```

```python
#separate stars and ratings column

def separate_ratings(data, column_name):
    ratings = []
    for value in data[column_name]:
        if value == 'Not rated yet':
            ratings.append(None)
            continue
        
        # Split the string into two parts based on the ' stars' substring
        parts = value.split(' stars')
        
      # Extract the second part and remove either ' ratings' or ' rating' substring
        rating_string = parts[1].replace(' ratings', '').replace(' rating', '')
        
        # Replace any commas in the string with nothing
        rating_string = rating_string.replace(',', '')
        
        # Convert the string to an integer
        number_of_ratings = int(rating_string)
        
        # Add the extracted number of ratings to the list
        ratings.append(number_of_ratings)
    
    # Add the list of ratings as a new column in the dataframe
    df['ratings'] = ratings
    
    return df

df = separate_ratings(df, 'stars')
print(df.head)
```

```python
#separate star value and if "Not rated yet" add in a 0 value


def f(row):
    if row == 'Not rated yet':
        return 0.0
    else:
        first_el=row.split(" ")[0]
        return first_el
df['stars']=df.stars.apply(f)
    
print(df)
```

```python
#convert time column to numerical int

def convert_time_to_minutes(row):
    time_string = row['time']
    time_parts = time_string.split()
    hours = 0
    minutes = 0
    for i in range(len(time_parts)):
        if time_parts[i] == 'hrs':
            hours = int(time_parts[i-1])
        elif time_parts[i] == 'mins':
            minutes = int(time_parts[i-1])
    total_minutes = hours * 60 + minutes
    return total_minutes

df['time'] = df.apply(convert_time_to_minutes, axis=1)
print(df)
```

```python
#reformat all dates

def convert_date_format(row):
    date_string = row['releasedate']
    if '/' in date_string:
        parts = date_string.split('/')
        month = parts[0]
        day = parts[1]
        year = parts[2]
    elif '-' in date_string:
        parts = date_string.split('-')
        day = parts[0]
        month = parts[1]
        year = parts[2]
    new_date = month + '/' + day + '/' + year
    return new_date

df['releasedate'] = df.apply(convert_date_format, axis=1)
print(df)
```

```python
#clean narrator row

def clean_narr(row):
    try:
        narr = row.split(":")[1].strip()
    except IndexError:
        narr = ""
    return narr

df['narrator'] = df['narrator'].apply(clean_narr)
print(df)
```

```python
#clean author row

def clean_auth(row):
    try:
        auth = row.split(":")[1].strip()
    except IndexError:
        auth = ""
    return auth

df['author'] = df['author'].apply(clean_auth)
print(df)
```

```python
def add_year_column(df, date_column_name, new_column_name):
    # Convert the date column to a pandas datetime object
    df[date_column_name] = pd.to_datetime(df[date_column_name], format='%m/%d/%y')
    
    # Extract the year from the datetime object and add it as a new column
    df[new_column_name] = df[date_column_name].dt.year
    

add_year_column(df, 'releasedate', 'releaseyear')
print(df)
```

```python
def add_month_column(df, date_column_name, new_column_name):
    # Convert the date column to a pandas datetime object
    df[date_column_name] = pd.to_datetime(df[date_column_name], format='%m/%d/%y')
    
    # Extract the year from the datetime object and add it as a new column
    df[new_column_name] = df[date_column_name].dt.month
    

add_month_column(df, 'releasedate', 'releasemonth')
print(df)
```
